## Web Collection Requirement \#1


```python
def web_parser1():
  url = 'https://www.reachhighscholars.org/scores_and_acceptance.html'
  resp = requests.get(url)
  soup = BeautifulSoup(resp.text, "html.parser")
  # print(soup)
  
  table = soup.find("table")
  rows = table.find_all("tr")

  headers = rows[0].text.strip('\n').replace('-', '\n\n').split('\n\n')
  headers[1] = headers[1] + 'th Percentile'
  headers[2] = 'SAT ' + headers[2]
  df = pd.DataFrame(columns = headers)
  # print(rows)
  rows = rows[1:]
  # print(rows)
  for row in rows:
    data = row.text.strip('\n').replace('-', '\n\n').split('\n\n')
    data = [i.strip() for i in data] #necessary in order to make sure that there is valuable data in the line and not just empty

    # print(data)
    if data != ['','',''] and len(data) == 4:
      df.loc[len(df)] = data
    elif(len(data) == 5):
      data = [data[0] + '-' + data[1], data[2], data[3]]
  # print(table)  
  df = df.astype({'SAT 25th Percentile':'int', 'SAT 75th Percentile': 'int'})

  df['SAT Average'] = (df['SAT 25th Percentile'] + df['SAT 75th Percentile']) / 2
  df.loc[106, '% of Applicants Accepted'] = int(df.iloc[106]['% of Applicants Accepted']) / 10
  df.to_csv("WebRequirement1.csv")

  return df
  # pass

############ Function Call ############
web_parser1()
```

### Insight 1 Explanation

Insight 1 was done to show the correlation between the percentage of acceptance applicants and the SAT average at each college. This insight corresponds to Data visualisation 1, which will be shown later but is a scatterplot with a trendline showing the correlation between the percentage of acceptance applicants and the SAT average at each college. As seen in this insight, the coefficient of determinant, also known as the R squared value, was approximately 0.7325, indicating a fairly strong correlation between the 2 variables.

```python
def insight1():
  df = pd.read_csv("WebRequirement1.csv", index_col=0)
  correlation = df['SAT Average'].corr(df['% of Applicants Accepted'])
  correlation_squared = correlation ** 2
  return f'The coefficient of determination (R squared value): ' + str(correlation_squared)
  # pass


############ Function Call ############
insight1()
```

```python
def visual1():
  df = pd.read_csv("WebRequirement1.csv", index_col=0)
  fig = px.scatter(df, x = '% of Applicants Accepted',
        y = 'SAT Average',
        title = 'Average SAT vs. % of Applicants Accepted',
        trendline = 'ols')
  fig.show()

  
  model = px.get_trendline_results(fig)
  alpha = model.iloc[0]["px_fit_results"].params[0]
  beta = model.iloc[0]["px_fit_results"].params[1]
  print(f'This is Visualization 1 which shows the relationship between the average SAT score for students at the school and the percentage of students accepted.\n')
  print(f'y-intercept: {alpha}')
  print(f'slope: {beta}')
  # return model
  pass

############ Function Call ############
visual1()
```

### Visualization 3 Explanation
